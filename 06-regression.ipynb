```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('csv_files/salaryData.csv')
dataset=dataset.dropna()
dataset.to_csv('new_dataset.csv')
```

```python
# d=all(dataset["Years of Experience"].isnull().values==True)
# d
# if any(dataset["Years of Experience"].isnull()==False):
#       print(dataset["Years of Experience"])
# dataset
```

```python
dataset = dataset.iloc[:,4:]
dataset
```

```python
X = dataset.iloc[:,0:-1].values
X
```

```python
Y = dataset.iloc[:,1:].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 1/3,random_state = 0)
```

```python
len(X_train)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
plt.scatter(X_train,Y_train,color = 'red')
plt.plot(X_train,model.predict(X_train),color = 'blue')
plt.show()
```

```python
plt.scatter(X_test,Y_test,color = 'red')
plt.plot(X_train,model.predict(X_train),color = 'blue')
plt.show()
```

```python
exp = float(input("Enter years of experience = "))
exp = np.array([[exp]])
predicted_salary = model.predict(exp)[0][0]
print("Your predicted_salary = ",predicted_salary)
```

```python
# for testing data
predicted_salary = model.predict(X_test)
print("Your predicted_salary = ",predicted_salary)
```
